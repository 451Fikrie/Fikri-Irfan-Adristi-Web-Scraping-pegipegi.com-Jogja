# 1.Scraping pegipegi.com Jogja

```python
#Import packages
import requests
import pandas as pd
from bs4 import BeautifulSoup
from google.colab import files
import csv
import time

#Entry coding
data_hotel=[]
counter_page = 0
#range diisi sebanyak page website
for page in range(1,2):
  counter_page=+1
  print('scraping page :',page)
  url = f'https://www.pegipegi.com/hotel/jogja/{page}.html/'
  headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36', "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

  req = requests.get(url, headers = headers)
  soup = BeautifulSoup(req.content, 'html.parser')
  items = soup.find_all('div','listContent')
    
  for it in items:
       #nama hotel
       try: 
        nama = it.find('div','title').text.replace('\n','').strip()
       except:
        nama= ''

       #Alamat hotel
       try:
        alamat = it.find('div','address').text.replace('\n','').strip().replace('Tampilkan Peta','')
       except :
        alamat=''
       
       #Bintang hotel
       try:
         star = len(it.find('div','title').find_all('i','fa fa-star'))
       except:
         star=''
       
       #Fitur wifi
       try :
        fasilitas = it.find('div','fiturText').text.replace('\n','').strip()
       except:
         fasilitas = ''
       
       #Fasilitas hotel tapi masih blm dapat
       fasum=soup.find_all('div','facility-highlight')
       for i in fasum:
         try:
           data=i.find_all('data-original-title').text.replace('\n','').strip()
         except:
           data=''
       
       #Rating hotel
       try:
          rating = it.find('div','ratingRight').find('span').text.replace('\n','').strip()
       except:
          rating = ''
       
       #Impresi hotel
       try:
          impresi = it.find('div','ratingRight').find_all('span')[1].text.replace('\n','')[:-1]
       except:
          impresi = ''

       #Jumlah Ulasan 
       try :
         jml_ulasan = it.find('div','ratingRight').find_all('span')[2].text.replace('\n','')[:-1].replace('(','')
       except:
         jml_ulasan =''
        
       try:
         #harga sblm diskon
         harga_normal = it.find('div','normalPrice').find('span').text.replace('\n','').strip().replace('Rp','')
       except :
         harga_normal =''

       try : 
        #hemat berapa persen
         hemat = it.find('div','normalPrice').find('b').text.replace('\n','').strip()
       except:
         hemat =''

       try:
         #tambahan diskon
         hemat_tambahan = it.find('div','discount-additional').text.replace('\n','').replace('+','').strip()
       except:
         hemat_tambahan = ''

       try :
         #harga setelah diskon
         harga_diskon = it.find('div', 'diskonPrice').text.replace('\n','').strip().replace('Rp','')
       except:
         harga_diskon='' 

       try:  
        #penawaran untuk hari raya
        limited_offer = it.find('div','note').text.strip()
       except :
        limited_offer = ''

       info_hotel={
              'nama' : nama,
              'star':star,
              'fasilitas':fasilitas,
              'fasum':fasum,
              'alamat' : alamat,
              'rating' : rating,
              'impresi': impresi,
              'jml_ulasan':jml_ulasan,
              'harga_normal': harga_normal,
              'hemat': hemat,
              'hemat_tambahan' : hemat_tambahan,
              'harga_diskon' : harga_diskon,
              'limited_offer': limited_offer
          }
       data_hotel.append(info_hotel)
       #data_hotel.append([nama, alamat, rating, impresi, harga_normal, hemat, harga_diskon,limited_offer])

#Pembuatan dataframe
df=pd.DataFrame(data_hotel)
df.head()
```
