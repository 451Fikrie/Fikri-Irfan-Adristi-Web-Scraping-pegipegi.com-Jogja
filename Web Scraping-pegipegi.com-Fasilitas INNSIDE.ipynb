```python
from bs4 import BeautifulSoup #digunakan untuk web scraping
import requests #digunakan untuk mendapatkan/mengakses/mendownload Webpage
import pandas as pd
```

## 2.Mendapatkan Webpage dari Link

```python
# Permintaan mendapatkan Webpage yang akan mengunduh konten HTML dari halaman web
getRequest = requests.get("https://www.pegipegi.com/hotel/jogja/innside_by_melia_yogyakarta_931370/?stayYear=2022&stayMonth=4&stayDay=22&stayCount=1&adultNum=2&autocomplete=jogja&roomCount=1")
getRequest
```

```python
# Menampilkan kode status dari permintaan pengunduhan konten HTML dari Webpage
getRequest.status_code
```

<br> 200 => OK, sukses

## 3.Melihat Secara General Content HTML yang didapatkan

```python
getRequest.text
```

Agar tampilan content menjadi lebih rapi, kita akan menggunakan BeautifulSoup constructor dan bantuan HTML parser.

```python
# Agar hasil terlihat lebih rapi dan dapat dibaca
soup = BeautifulSoup(getRequest.text, 'html.parser')
print(soup.prettify()) #fungsi prettify() membuat hasil dapat dibaca
```
